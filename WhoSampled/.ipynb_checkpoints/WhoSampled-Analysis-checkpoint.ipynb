```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import spearmanr
import seaborn as sns
from scipy import stats
```

```python
df = pd.read_csv('results.csv')
df = df.drop(columns = ['id'])
df = df.drop_duplicates()
```

```python
test = df[df['Sampled Times'] != -1]
```

# Sampled Times, How many songs sample the song

```python
#Spearman
sampledSongs = df[df['Sampled Times'] != -1]
sampledSongs = sampledSongs.sort_values(by=['Sampled Times'])
plt.figure(figsize=(12,8))
corr = sampledSongs.drop(columns=['release_date', 'year', 'CoveredTimes','Remix Times', 'Remix of ', 'Sample of', 'Cover Of']).corr(method='spearman')
CM = sns.heatmap(corr, annot=True, annot_kws={"size": 12})
plt.title('Spearman HeatMap, Sampled Times Raw Data')
plt.show()
corr
```

```python
#Average metrics by times sampled w/ spearman
sampledMeans = sampledSongs.groupby(['Sampled Times']).mean()
sampledMeans = sampledMeans.reset_index()
plt.figure(figsize=(12,8))
corr = sampledMeans.drop(columns=['CoveredTimes', 'year', 'Remix Times', 'Remix of ', 'Sample of', 'Cover Of']).corr(method='spearman')
CM = sns.heatmap(corr, annot=True, annot_kws={"size": 12})
plt.title('Spearman HeatMap, Sampled Times Averaged Data')
plt.show()
```

```python
#Pearson
sampledSongs = df[df['Sampled Times'] != -1]
sampledSongs = sampledSongs.sort_values(by=['Sampled Times'])
plt.figure(figsize=(12,8))
corr = sampledSongs.drop(columns=['release_date', 'year', 'CoveredTimes','Remix Times', 'Remix of ', 'Sample of', 'Cover Of']).corr(method='pearson')
CM = sns.heatmap(corr, annot=True, annot_kws={"size": 12})
plt.title('Spearman HeatMap, Sampled Times Raw Data')
#plt.show()
corr
```

```python
#Valence
plt.title("Average valence for each number of time sampled")
plt.plot(sampledMeans['Sampled Times'].values, sampledMeans.valence.values)
plt.xlabel("Times Sampled")
plt.ylabel("tempo")
plt.show()
plt.title("Valence to Times Sampled")
plt.scatter(sampledSongs['Sampled Times'].values, sampledSongs.valence.values)
plt.xlabel("Times Sampled")
plt.ylabel("tempo")
plt.show()
#Energy
plt.title("Average Evergy for each number of time sampled")
plt.plot(sampledMeans['Sampled Times'].values, sampledMeans.energy.values)
plt.xlabel("Times Sampled")
plt.ylabel("tempo")
plt.show()
plt.title("Energy to Times Sampled")
plt.scatter(sampledSongs['Sampled Times'].values, sampledSongs.energy.values)
plt.xlabel("Times Covered")
plt.ylabel("tempo")
plt.show()
#Loudness
plt.title("Average loudness for each number of time sampled")
plt.plot(sampledMeans['Sampled Times'].values, sampledMeans.loudness.values)
plt.xlabel("Times Covered")
plt.ylabel("tempo")
plt.show()
plt.title("loudness to Times Sampled")
plt.scatter(sampledSongs['Sampled Times'].values, sampledSongs.loudness.values)
plt.xlabel("Times Covered")
plt.ylabel("tempo")
plt.show()
```

# Comparisons of Songs That have have records in whosampled categories vs those who do now

```python
testcols = df.drop(columns=['name', 'artists', 'release_date', 'year', 'Sampled Times', 'CoveredTimes', 'Remix Times', 'Remix of ', 'Sample of', 'Cover Of']).columns
testcols
```

<h3> Comparison of songs containing sampled and not

```python
sampleOfComp = pd.DataFrame(columns = ['Metric','Contains Sample(s)', 'Sample Free'])
data_to_append = {}
for i in testcols:
    if i == 'key' or i == 'mode':
        data_to_append['Metric'] = i
        data_to_append['Contains Sample(s)'] = stats.mode(df[df['Sample of']!=-1][i].values)
        data_to_append['Sample Free'] = stats.mode(df[df['Sample of']==-1][i].values)
    else:
        data_to_append['Metric'] = i
        data_to_append['Contains Sample(s)'] = df[df['Sample of']!=-1][i].values.mean()
        data_to_append['Sample Free'] = df[df['Sample of']==-1][i].values.mean()
    sampleOfComp = sampleOfComp.append(data_to_append, ignore_index=True)
sampleOfComp
```

<h3>Comparison of songs that have been sampled and not

```python
sampledComp = pd.DataFrame(columns = ['Metric','Has Been Sampled', 'Not been sampled'])
data_to_append = {}
for i in testcols:
    if i == 'key' or i == 'mode':
        data_to_append['Metric'] = i
        data_to_append['Has Been Sampled'] = stats.mode(df[df['Sampled Times']!=-1][i].values)
        data_to_append['Not been sampled'] = stats.mode(df[df['Sampled Times']==-1][i].values)
    else:
        data_to_append['Metric'] = i
        data_to_append['Has Been Sampled'] = df[df['Sampled Times']!=-1][i].values.mean()
        data_to_append['Not been sampled'] = df[df['Sampled Times']==-1][i].values.mean()
    sampledComp = sampledComp.append(data_to_append, ignore_index=True)
sampledComp
```

<h3> Comparisons of songs that have been covered and not

```python
coveredComp = pd.DataFrame(columns = ['Metric','Has Been covered', 'Not been covered'])
data_to_append = {}
for i in testcols:
    if i == 'key' or i == 'mode':
        data_to_append['Metric'] = i
        data_to_append['Has Been covered'] = stats.mode(df[df['CoveredTimes']!=-1][i].values)
        data_to_append['Not been covered'] = stats.mode(df[df['CoveredTimes']==-1][i].values)
    else:
        data_to_append['Metric'] = i
        data_to_append['Has Been covered'] = df[df['CoveredTimes']!=-1][i].values.mean()
        data_to_append['Not been covered'] = df[df['CoveredTimes']==-1][i].values.mean()
    coveredComp = coveredComp.append(data_to_append, ignore_index=True)
coveredComp
```
