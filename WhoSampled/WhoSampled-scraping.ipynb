```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
import time
```

```python
# Set up data set to incorperate data from who sampled
df = pd.read_csv('data.csv')
df = df.sort_values(by=['year'], ascending=False)
df['Sampled Times'] = -1
df['Covered Times'] = -1
df['Remixed Times'] = -1
df['Remix of'] = -1
df['Sample of'] = -1
df['Cover of'] = -1
results = pd.DataFrame(columns = df.columns)
df = df.set_index('id')

hds = {'Host': 'www.whosampled.com', 'User-Agent': 'Custom'}
host = 'https://www.whosampled.com/'
success = 0
failure = 0

```

```python
for i in range(15305, 100000):
    #extract useable information from dataset
    id = df.index[i]
    artists = df['artists'][i]
    artists = artists[1:len(artists)-1]
    artist = artists.split(", ")[0]
    artist = artists[1:len(artist)-1]
    name = df['name'][i]
    uname = name
    #alter names to better fit whosampled url policies
    if " - " in uname:
        uname = uname.split(" - ")[0]
    if "(Live)" in uname:
        uname = uname.replace(" (Live)", "")
    if "(feat" in uname:
        uname.split("(feat")[0]
    if "[feat" in uname:
        uname.split("[feat")[0]
    uname = uname.replace(" ", "-")
    artist = artist.replace(" ", "-")
    
    
    #get the data
    url = host+artist+"/"+uname
    page = requests.get(url, headers = hds)
    soup = bs(page.text, 'html.parser')
    print(str(i) + ", " +page.url)
    #No whosampled data available
    if len(soup.find_all(class_='error')) == 1:
        #print('error')
        failure += 1
        failed.append((name, artist))
    #Possible data available, look fo specific needed data in response  
    else:
        try:
            for r in soup.find_all(class_='section-header-title'):
                #print(r.contents[0].split())
                if "Was sampled in " in r.contents[0]:
                    df.at[id, 'Sampled Times'] = [int(i) for i in r.contents[0].split() if i.isdigit()][0]

                if "Was covered in " in r.contents[0]:
                    df.at[id, 'Covered Times'] = [int(i) for i in r.contents[0].split() if i.isdigit()][0]

                if "Was remixed in " in r.contents[0]:
                    df.at[id, 'Remixed Times'] = [int(i) for i in r.contents[0].split() if i.isdigit()][0]

                if "Is a remix of" in r.contents[0]:
                    df.at[id, 'Remix of'] = [int(i) for i in r.contents[0].split() if i.isdigit()][0]

                if "Contains samples of " in r.contents[0]:
                    df.at[id, 'Sample of'] = [int(i) for i in r.contents[0].split() if i.isdigit()][0]

                if "Is a cover of" in r.contents[0]:
                    df.at[id, 'Cover of'] = [int(i) for i in r.contents[0].split() if i.isdigit()][0]
        except:
            print("the bug")
        df.loc[[id]].to_csv('results.csv', header = False, mode='a')
        
                
    time.sleep(4)
    
```
