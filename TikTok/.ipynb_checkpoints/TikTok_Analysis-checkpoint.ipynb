#**Is There a Correlation Between the Descriptive Metrics of Songs Provided by Spotify and Song Popularity on Various Platforms?**


## **1.0 Overview:**


---
###**1.1 Analyzing Popularity of Trending TikTok Songs Against Spotify Metrics**

The application Spotify contains a wide range of songs, intended to fit any user’s music taste and preferred genre. To determine whether the future popularity of these songs can be predicted based on their inherent characteristics, correlations must be investigated between a known set of song metrics and song popularity at various times on several platforms.

Songs from the chosen Spotify dataset (https://www.kaggle.com/yamaerenay/spotify-dataset-19212020-160k-tracks) are described using the following metrics:

Numerical:

- acousticness (Ranges from 0 to 1)
- danceability (Ranges from 0 to 1)
- energy (Ranges from 0 to 1)
- duration_ms (Integer typically ranging from 200k to 300k)
- instrumentalness (Ranges from 0 to 1)
- valence (Ranges from 0 to 1)
- popularity (Ranges from 0 to 100)
- tempo (Float typically ranging from 50 to 150)
- liveness (Ranges from 0 to 1)
- loudness (Float typically ranging from -60 to 0)
- speechiness (Ranges from 0 to 1)
- year (Ranges from 1921 to 2020)

Dummy:

- mode (0 = Minor, 1 = Major)
- explicit (0 = No explicit content, 1 = Explicit content)

Categorical:

- key (All keys on octave encoded as values ranging from 0 to 11, starting on C as 0, C# as 1 and so on…)
- artists (List of artists mentioned)
- release_date (Date of release mostly in yyyy-mm-dd format, however precision of date may vary)
- name (Name of the song)



---

Questions to answer:

- What correlations, if any, exist between a song's TikTok popularity and its Spotify metrics?
- What Spotify musical factor was maximized among the most popular songs on TikTok?

This analysis will measure the popularity of a song on TikTok via four quantities, gained from the TikTok API: the diggCount (the number of times a video has been liked), the shareCount (the number of times a video has been shared), the playCount (the number of times a video has been played), and the commentCount (the number of comments a video has).

Originally, the author planned to gather data from TikTok about each song in the above Spotify dataset, looking at a representative sample of videos using each song to gain a picture of how each song impacted video performance. However, this approach was not possible due to the limitations of the TikTok API. The API only allows music look-up via a numerical ID, which is not easily accessible for any given song. Thus, an alternative approach was taken: the top 20,000 trending videos on TikTok at the time of analysis were scraped, then those videos which used songs quantified in the Spotify dataset were matched to their Spotify metrics. This method has no doubt introduced bias, as it only looks at those videos which were already trending on TikTok. The method by which the TikTok algorithm determines what is "trending" is also unknown.

## **2.0 Gathering and Preparing the Data**


---



```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
import numpy as np
import seaborn as sns
```

## **2.1 Spotify**

```python
spotify_df = pd.read_csv('data.csv')
spotify_df = pd.DataFrame(spotify_df, columns=["name","artist","year","popularity","acousticness",\
    "danceability","energy","duration_ms","instrumentalness","valence","tempo","liveness","loudness","speechiness"])

#Remove [] and '' from Spotify artist names
spotify_df['artist'] = spotify_df['artist'].map(lambda x: x.lstrip("['").rstrip("']"))

#Remove duplicates with the same song name
spotify_df = spotify_df.drop_duplicates(subset=['name'])
spotify_df
```

## **3.0 Analysis**

```python
tiktok_df = pd.read_csv('TikTok_data.csv') #Change this to the name of your trends file, if scraping has been run again
tiktok_df = tiktok_df[tiktok_df['musicMeta.musicOriginal'] != True] #Filter out videos with original sound
tiktok_df = tiktok_df[["musicMeta.musicName","musicMeta.musicAuthor","diggCount","shareCount","playCount","commentCount"]] #Extract relevant columns

#Merge TikTok data with Spotify data
merge_df = pd.merge(tiktok_df, spotify_df, left_on = ['musicMeta.musicName','musicMeta.musicAuthor'], right_on = ['name','artist'])
merge_df
```

```python
#Remove popularity metric outliers more than 3 standard deviations from mean
for columnName in ['diggCount','shareCount','playCount','commentCount']:
    merge_df = merge_df[(np.abs(stats.zscore(merge_df[columnName])) < 3)]
```

```python
#TikTok popularity metrics
digg = merge_df['diggCount']
share = merge_df['shareCount']
play = merge_df['playCount']
comment = merge_df['commentCount']

#Spotify song metrics
acousticness = merge_df['acousticness']
danceability = merge_df['danceability']
energy = merge_df['energy']
instrumentalness = merge_df['instrumentalness']
liveness = merge_df['liveness']
loudness = merge_df['loudness']
speechiness = merge_df['speechiness']
tempo = merge_df['tempo']
valence = merge_df['valence']
```

```python
# Calculate the correlation matrix between the TikTok popularity and the Spotify metrics
fig, ax = plt.subplots(figsize=(12,12))
corr_matrix = merge_df.corr()
sns.set(font_scale=0.8)
sns.heatmap(corr_matrix, vmax = .8, square = True, annot = True)
```

```python
#Scatter plot creation
spotify_metrics = ['acousticness','danceability','energy','instrumentalness','liveness','loudness','speechiness','tempo','valence']
tiktok_metrics = ['diggCount','shareCount','playCount','commentCount']
colours = ['red','blue','green','purple']

plt.figure(figsize = (30,45))
i = 1
for spotify in spotify_metrics:
  for tiktok,colour in zip(tiktok_metrics,colours):
    plt.subplot(9,4,i); 
    plt.scatter(merge_df[spotify],merge_df[tiktok],s=2,\
                label=f'Correlation = {np.round(np.corrcoef(merge_df[spotify],merge_df[tiktok])[0,1], 2)}', c=colour);
    plt.xlabel(spotify); plt.ylabel(tiktok)
    plt.title(f'{spotify} vs. {tiktok}')
    plt.legend(loc="upper right")
    i+=1

plt.show()
```

From the correlation matrix above, the playCount was identified as a TikTok metric of interest. This was highly correlated with diggCount (0.86 correlation) and requires the lowest effort from TikTok users of all the popularity metrics. Thus, the Spotify metrics were plotted separately vs. the playCount, shown below.

```python
spotify_metrics = ['acousticness','danceability','energy','instrumentalness','liveness','loudness','speechiness','tempo','valence']
plt.figure(figsize = (65,65))
i = 1
for spotify in spotify_metrics:
  x = merge_df[spotify]
  y = merge_df['playCount']
  p_corr = np.round(np.corrcoef(x,y)[0,1], 2)
  plt.subplot(3,3,i); 
  plt.scatter(x,y,s=80,label=f'Pearson \nCorrelation = {p_corr}')
  plt.legend(loc="upper right",fontsize=80)                                                                      
  plt.xlabel(spotify,fontsize=80); 
  plt.ylabel('Play Count',fontsize=80)
  plt.title(f'{spotify} vs. playCount',fontsize=80)
  m, b = np.polyfit(x, y, 1);
  plt.plot(x, m*x+b,c='red',lw=10.0)
  i+=1

plt.show()
```

## **4.0 Conclusions**


---
In conclusion, there does not appear to exist any signficant correlations between the investigated Spotify metrics and the popularity of trending videos on TikTok, as measured using the diggCount, shareCount, playCount, or commentCount. These results may be biased due to the extraction of only trending TikToks.
